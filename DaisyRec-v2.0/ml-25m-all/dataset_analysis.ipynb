```python
import pandas as pd
import numpy as np
```

```python
years =np.arange(2015,2019,1)
files = ['all', 'rem15', 'srem15', 'brem15']
for year in years:
    for file in files:
        if file == 'srem15':
            file_name = 'S'+str(year)+'rem15.csv'
        elif file == 'brem15':
            file_name = 'B'+str(year)+'rem15.csv'

        else:
            file_name = str(year)+file+'.csv'
        data = pd.read_csv(file_name, header=0,sep=',')
        print(data.head())
        print(len(data))
```
